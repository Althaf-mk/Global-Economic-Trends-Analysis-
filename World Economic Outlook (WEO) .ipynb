<h1 style="font-size:10x; color:skyblue;">
2.4 Creating Derived Columns</h1>

```python
years = [col for col in df.columns if str(col).isdigit()]

# Create average column
df["Average_Value"] = df[years].mean(axis=1)

print(df[["COUNTRY", "Average_Value"]].head())
print("Successfully created")
```

<div style="background-color:#d4edda; padding:10px;">
<p> <b><h3>Insight</h3></b></p>

This code creates a new column called `Average_Value` in the dataset by calculating the average of selected yearly columns stored in `years`. The `mean(axis=1)` function computes the row-wise average, meaning it takes all yearly values for each record and finds their average value.

The newly created `Average_Value` column helps summarize multiple yearly economic values into a single representative value, making it easier to compare countries and perform further analysis. Finally, `print(df[["COUNTRY", "Average_Value"]].head())` displays the first five rows containing the country name and its calculated average value.



</div>

<div style="background-color:#d4edda; padding:10px;">
<p> <b><h3>Insight</h3></b></p>
The violin plot was used to visualize the distribution of the Average_Value column. Initially, the plot appeared compressed because the dataset contained a few extremely large values (outliers), causing a highly right-skewed distribution. To address this issue, a logarithmic transformation (log1p) was applied to the data before plotting.

The log transformation reduces the impact of extreme values and spreads the data more evenly across the scale. This makes it easier to observe the underlying distribution, density, and variability of the majority of observations. The resulting violin plot provides a clearer representation of the dataset and helps identify concentration patterns that were hidden in the original scale.

<h1 style="font-size:10x; color:skyblue;">
4.21 Radar Chart</h1>

```python
from math import pi

country = df.groupby("COUNTRY")[years].mean().iloc[0]

values = country.values.tolist()
values += values[:1]

angles = [n / float(len(years)) * 2 * pi for n in range(len(years))]
angles += angles[:1]

plt.figure(figsize=(8,8))
ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1], years)
ax.plot(angles, values)
ax.fill(angles, values, alpha=0.25)

plt.title("Country Economic Profile")
plt.show()
```

<div style="background-color:#d4edda; padding:10px;">
<p> <b><h3>Insight</h3></b></p>
The radar chart visualizes the economic profile of a country across multiple years. Each axis represents a year, while the distance from the center indicates the average economic value for that period.

This chart provides a unique way to observe the consistency and variation of economic performance over time. Larger areas generally indicate higher economic values, while irregular shapes may highlight periods of growth, decline, or fluctuation.


</div>
